```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('combined_output.csv')
df
```

```python
new_order = ['ID','Title','Difficulty','Leetcode Question Link','Acceptance','Frequency']
new_order
```

```python
df = df[new_order]
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
duplicates = df[df.duplicated()]
print(duplicates)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Difficulty'] = le.fit_transform(df['Difficulty'])
def convert_percentage_to_float(value):
    if isinstance(value, str) and '%' in value:
        return float(value.replace('%', '')) / 100.0
    return float(value)

df['Frequency'] = df['Frequency'].apply(convert_percentage_to_float)
df['Acceptance'] = df['Acceptance'].apply(convert_percentage_to_float)
```

```python
X = df[['Difficulty']]
Y = df[['Frequency', 'Acceptance']]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
rf = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
from sklearn.multioutput import MultiOutputRegressor
multi_target_model = MultiOutputRegressor(rf)
```

```python
multi_target_model.fit(X_train, Y_train)
```

```python
Y_pred = multi_target_model.predict(X_test)
```

```python
mse_freq = mean_squared_error(Y_test['Frequency'], Y_pred[:, 0])
mse_acc = mean_squared_error(Y_test['Acceptance'], Y_pred[:, 1])
```

```python
print(mse_acc)
print(mse_freq)
```

```python
import pickle

# Assume 'model' is your trained model
model = multi_target_model  # Replace this with your actual model

# Save the model to a file
with open('model.pkl', 'wb') as file:
    pickle.dump(model, file)

print("Model saved to 'model.pkl'")
```
